# VacinaBR-PNI - Exploratory Analysis

Notebook de exemplo para explorar os CSVs analiticos gerados pela pipeline.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path('..')
ANALYTICS_DIR = ROOT / 'data' / 'analytics'

def read_csv(name):
    return pd.read_csv(ANALYTICS_DIR / name)
```

## Doses por mes

Usa `monthly_vaccination_summary.csv`, que ja esta agregado por ano e mes.

```python
monthly = read_csv('monthly_vaccination_summary.csv')
monthly['periodo'] = monthly['ano_vacinacao'].astype(str) + '-' + monthly['mes_vacinacao'].astype(str).str.zfill(2)
monthly
```

```python
ax = monthly.plot.bar(x='periodo', y='doses_aplicadas', legend=False, figsize=(10, 4))
ax.set_xlabel('Periodo')
ax.set_ylabel('Doses aplicadas')
ax.set_title('Doses aplicadas por mes')
plt.tight_layout()
```

## Principais vacinas

Usa `vaccine_type_summary.csv`, agora com `sg_vacina` e `descricao_vacina`.

```python
vaccines = read_csv('vaccine_type_summary.csv')
vaccines.head(15)
```

```python
top_vaccines = vaccines.head(10).sort_values('doses_aplicadas')
ax = top_vaccines.plot.barh(x='descricao_vacina', y='doses_aplicadas', legend=False, figsize=(10, 5))
ax.set_xlabel('Doses aplicadas')
ax.set_ylabel('Vacina')
ax.set_title('Top 10 vacinas por doses aplicadas')
plt.tight_layout()
```
