# VacinaBR-PNI - Data Quality Report

Notebook de exemplo para analisar completude, validade de idade e validade documental por grupo.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path('..')
ANALYTICS_DIR = ROOT / 'data' / 'analytics'

def read_csv(name):
    return pd.read_csv(ANALYTICS_DIR / name)
```

## Qualidade por UF

Responde perguntas como: quais UFs concentram mais registros incompletos?

```python
quality_state = read_csv('quality_by_state.csv')
quality_state['registros_incompletos'] = quality_state['total_registros'] - quality_state['registros_completos']
quality_state.sort_values('registros_incompletos', ascending=False).head(15)
```

```python
plot_state = quality_state.sort_values('pct_completude').head(15)
ax = plot_state.plot.barh(x='uf_paciente', y='pct_completude', legend=False, figsize=(8, 5))
ax.set_xlabel('% completude')
ax.set_ylabel('UF')
ax.set_title('UFs com menor completude')
plt.tight_layout()
```

## Qualidade por vacina

Ajuda a identificar se algum tipo de vacina concentra mais registros incompletos ou documentos invalidos.

```python
quality_vaccine = read_csv('quality_by_vaccine.csv')
quality_vaccine.sort_values(['pct_completude', 'total_registros'], ascending=[True, False]).head(20)
```

## Evolucao mensal da qualidade

Permite verificar se a qualidade melhora ou piora ao longo do ano.

```python
quality_month = read_csv('quality_by_month.csv')
quality_month['periodo'] = quality_month['ano_vacinacao'].astype(str) + '-' + quality_month['mes_vacinacao'].astype(str).str.zfill(2)
quality_month[['periodo', 'pct_completude', 'pct_idade_valida', 'pct_documento_valido']]
```

```python
ax = quality_month.plot.line(
    x='periodo',
    y=['pct_completude', 'pct_idade_valida', 'pct_documento_valido'],
    marker='o',
    figsize=(10, 4),
)
ax.set_xlabel('Periodo')
ax.set_ylabel('%')
ax.set_title('Indicadores de qualidade por mes')
plt.tight_layout()
```
